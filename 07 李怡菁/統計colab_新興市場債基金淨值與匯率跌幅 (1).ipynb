```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming your data is in a DataFrame named 'data' with columns '淨值漲跌幅' and '匯率漲跌幅'
data = pd.read_csv('/content/drive/MyDrive/@NYCU/112 統計學/112統計學-Colab Notebooks/期末報告資料/統計colab-新興市場基金匯率淨值跌幅r7.csv')
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib.font_manager import fontManager

fontManager.addfont('/content/drive/MyDrive/@NYCU/112 統計學/112統計學-Colab Notebooks/期末報告資料/TaipeiSansTCBeta-Regular (2).ttf')

mpl.rc('font', family='Taipei Sans TC Beta')
# Scatter plot with regression line
plt.figure(figsize=(8, 10))
sns.regplot(x='匯率漲跌幅', y='淨值漲跌幅', data=data)

# Adding labels and title
plt.xlabel('匯率漲跌幅')
plt.ylabel('淨值漲跌幅')
plt.title('淨值漲跌幅和匯率漲跌幅的回歸線')
# Display the plot
plt.show()

import pandas as pd
import statsmodels.api as sm

# Add a constant term for the intercept
data['const'] = 1
# Specify the independent variable and dependent variable
X = data[['const', '匯率漲跌幅']]
y = data['淨值漲跌幅']
# Fit the regression model
model = sm.OLS(y, X).fit()

# Display regression results
print(model.summary())

# Get the coefficients
intercept, slope = model.params['const'], model.params['匯率漲跌幅']

# Display the regression equation
print(f"Regression Equation: 淨值漲跌幅 = {intercept:.4f} + {slope:.4f} * 匯率漲跌幅")
```
