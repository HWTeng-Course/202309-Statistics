```python
import pandas as pd
import matplotlib.pyplot as plt

# 假设数据保存在一个名为 data.csv 的 CSV 文件中
data = pd.read_csv('data.csv')

# 将 'year' 列调整为字符串格式
data['year'] = (data['year'].str.slice(0, 3).astype(int) + 1911).astype(str) + '-' + data['year'].str.slice(3)

# 拆分 'year' 列并创建新的列 'quarter'
data[['year', 'quarter']] = data['year'].str.split('Q', expand=True)

# 转换为日期格式
data['date'] = pd.to_datetime(data['year'] + '-' + data['quarter'])

# 绘制散点图
plt.figure(figsize=(10, 6))
plt.scatter(data['date'], data['consumer_price'], label='Consumer Price Index', marker='o')
plt.scatter(data['date'], data['housing_price'], label='Housing Price Index', marker='x')
plt.title('Consumer Price Index vs. Housing Price Index Over Time')
plt.xlabel('Year')
plt.ylabel('Index Value')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 假设你的数据保存在一个名为 data.csv 的 CSV 文件中
data = pd.read_csv('data.csv')

# 绘制 scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(data['consumer_price'], data['housing_price'], marker='o', color='blue')
plt.title('Scatter Plot: Consumer Price index vs. Housing Price index')
plt.xlabel('Consumer Price Index')
plt.ylabel('Housing Price Index')
plt.grid(True)

# 使用 polyfit 进行线性拟合，返回的 coef 包含截距和斜率
coef = np.polyfit(data['consumer_price'], data['housing_price'], 1)

# 提取截距和斜率
slope, intercept = coef

# 构建回归线的方程 y = a + bx
regression_line = f'y = {intercept:.2f} + {slope:.2f}x'

# 绘制回归线
plt.plot(data['consumer_price'], intercept + slope * data['consumer_price'], color='red', label=regression_line)
plt.legend()

# 调整坐标轴范围
plt.xlim(min(data['consumer_price']) - 1, max(data['consumer_price']) + 1)
plt.ylim(min(data['housing_price']) - 1, max(data['housing_price']) + 1)

plt.show()

# 输出回归线的截距和斜率
print(f'Intercept (a): {intercept:.2f}')
print(f'Slope (b): {slope:.2f}')
```
