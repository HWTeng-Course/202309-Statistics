```python
# Problem Formulation and Analysis


import matplotlib as mpl
import matplotlib.pyplot as plt
import csv
import numpy as np
import pandas as pd
import statistics
from matplotlib.font_manager import fontManager
from sklearn.linear_model import LinearRegression
from scipy import stats
import scipy.stats

# Colab 進行matplotlib繪圖時顯示繁體中文
# 下載台北思源黑體並命名taipei_sans_tc_beta.ttf，移至指定路徑
!wget -O TaipeiSansTCBeta-Regular.ttf https://drive.google.com/uc?id=1eGAsTN1HBpJAkeVM57_C7ccp7hbgSz3_&export=download
fontManager.addfont('TaipeiSansTCBeta-Regular.ttf')
mpl.rc('font', family='Taipei Sans TC Beta', size=15)

# 從GoogleDrive讀取csv檔案
data='/content/drive/MyDrive/CSV_原始檔/Jay_Chou_Album.csv'
album=pd.read_csv(data)

album_stat_bf_2008=album.head(8)
album_stat_af_2008=album.tail(7)

group1_data=album_stat_bf_2008['專輯銷量 (萬張)']
group2_data=album_stat_af_2008['專輯銷量 (萬張)']

print('')
print("\033[1;103;30m 4.Problem Formulation and Analysis \033[0m")
print("")

# t檢定(全部15張專輯)
t_statistic, p_value = stats.ttest_ind(group1_data, group2_data)
print('-----------------------------------------------------')
print('')
print("\033[1;36;6m [全部15張專輯, n=15, df=13] \033[0m")
print('')
print(f'T-statistic: {t_statistic}')
print(f'P-value: {p_value}')

# 進行統計檢定
critical=scipy.stats.t.ppf(q=1-.05/2,df=13)
print(f'T critical value: {critical}')
print('')
print('Determine T-test result by rejection region method :')
if t_statistic>critical:
  print("=> reject H0")
else:
  print("=> Do not reject H0")
print('')
print('Determine T-test result by P-value method :')
if p_value<.05:
  print("=> reject H0")
else:
  print("=> Do not reject H0")
print('')
print('-----------------------------------------------------')
print('')

# t檢定(扣除最後兩張數位專輯，全部13張專輯)
album_stat_af_2008_no_digital=album_stat_af_2008.head()
group3_data=album_stat_af_2008_no_digital['專輯銷量 (萬張)']

t_statistic, p_value = stats.ttest_ind(group1_data, group3_data)
print("\033[1;36;6m [扣除最後兩張數位專輯, n=13, df=11] \033[0m")
print('')
print(f'T-statistic: {t_statistic}')
print(f'P-value: {p_value}')

# 進行統計檢定
critical=scipy.stats.t.ppf(q=1-.05/2,df=11)
print(f'T critical value: {critical}')
print('')
print('Determine T-test result by rejection region method :')
if t_statistic>critical:
  print("=> reject H0")
else:
  print("=> Do not reject H0")
print('')
print('Determine T-test result by P-value method :')
if p_value<.05:
  print("=> reject H0")
else:
  print("=> Do not reject H0")
print('')
print('-----------------------------------------------------')
```

```python
import matplotlib as mpl
import csv
from matplotlib.font_manager import fontManager

#導入統計模型套件
import statsmodels.api as sm
from statsmodels.formula.api import ols

#導入Python數據處理套件
import numpy as np
import pandas as pd

#導入視覺化套件
import matplotlib.pyplot as plt

#導入Sklearn套件
#導入將數據集拆成訓練集與測試集的套件
from sklearn.model_selection import train_test_split

#導入迴歸模型套件
from sklearn.linear_model import LinearRegression


# Colab 進行matplotlib繪圖時顯示繁體中文
# 下載台北思源黑體並命名taipei_sans_tc_beta.ttf，移至指定路徑
!wget -O TaipeiSansTCBeta-Regular.ttf https://drive.google.com/uc?id=1eGAsTN1HBpJAkeVM57_C7ccp7hbgSz3_&export=download
fontManager.addfont('TaipeiSansTCBeta-Regular.ttf')
mpl.rc('font', family='Taipei Sans TC Beta', size=15)


# 另一種方式: 訓練迴歸模型
print('')
print('')
print("\033[1;36;6m[另一種方式] 將總數據集拆出訓練集，使用迴歸模型訓練，再拿訓練好的迴歸模型預測測試集資料的目標值\033[0m")
print('')
print('')


# 從GoogleDrive讀取csv檔案
data='/content/drive/MyDrive/CSV_原始檔/Jay_Chou_Spotify.csv'
Spotify=pd.read_csv(data,usecols=[0,1,2])
display(Spotify)   # 顯示讀取的資料內容
print('')
print('')


# 繪製散布圖
print('')
print("\033[1;36;6m散布圖\033[0m")
print("")
file=open("/content/drive/MyDrive/CSV_原始檔/Jay_Chou_Spotify.csv",encoding="utf-8-sig")
reader=csv.reader(file)
header=next(reader)

x=[]
y=[]
for row in reader:
  x.append(float(row[1]))
  y.append(float(row[2]))

plt.scatter(x,y)
plt.xlabel(header[1])
plt.ylabel(header[2])
plt.show()
print('')
print('')


# 導入數據集
dataset = pd.read_csv('/content/drive/MyDrive/CSV_原始檔/Jay_Chou_Spotify.csv')
X = dataset.iloc[:, 1].values.reshape(-1,1)
y = dataset.iloc[:,2].values

# 將數據集拆成訓練集與測試集
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)

# 訓練與建構迴歸模型
regressor = LinearRegression()
regressor.fit(X_train, y_train)

# 計算出截距值與係數值
w_0 = regressor.intercept_
w_1 = regressor.coef_

print('Interception : ', w_0)
print('Coeficient : ', w_1)

# 迴歸模型的準確度
score = regressor.score(X_test, y_test)
print('Score: ', score)
print('Accuracy: ' + str(score*100) + '%')

# 拿訓練好的迴歸模型預測測試集資料的目標值(依變數)
y_pred = regressor.predict(X_test)
print('Predict : ', y_pred)
print('')
print('')

# 視覺化迴歸模型與訓練集的關聯
plt.scatter(X_train, y_train, color = 'red')
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.title('專輯銷量 (百萬張) vs Spotify播放量 (億次) [訓練集]')
plt.xlabel("Spotify播放量 (億次)")
plt.ylabel("專輯銷量 (百萬張)")
plt.show()
print('')
print('')

# 視覺化迴歸模型與測試集的關聯
plt.scatter(X_test, y_test, color = 'red')
plt.plot(X_test, regressor.predict(X_test), color = 'blue')
plt.title('專輯銷量 (百萬張) vs Spotify播放量 (億次) [測試集]')
plt.xlabel("Spotify播放量 (億次)")
plt.ylabel("專輯銷量 (百萬張)")
plt.show()
```
